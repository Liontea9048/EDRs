```python

def multiplyingValues(data, workingType, mul = 1.e6):
    mcCode = [f"GEANT4_{workingType}", f"MCNP6_{workingType}", f"PHITS_{workingType}"]
    for key in data[mcCode]:
        nanlst = data[key].isna()
        for i in range(len(nanlst)):
            if not nanlst[i]:
                data[key][i] = data[key][i] * 1.e6
            else:
                pass

multiplyingValues(dfOp, "op")
multiplyingValues(dfSp, "sp")
multiplyingValues(dfSq, "sq")
multiplyingValues(dfBd, "bd")
```

```python
def getCoef(x, y, level, yerr):
    
    def func(x, a, b):
        return a + b * x
    
    if level == 1:
        newx = x[:10]
        newy = y[:10]
        newyerr = yerr[:10]
        
    elif level == 2:
        newx = x[13:]
        newy = y[13:]
        newyerr = yerr[13:]
        # print(newx)
        
    elif level == 3:
        newx = x
        newy = y
        newyerr = yerr
        
    logy = np.log(newy)
    
    popt, pcov = curve_fit(func, newx, logy, method="lm")
    
    # get r squared
    c = np.exp(popt[0]) * np.exp(popt[1] * newx)
    correlation_matrix = np.corrcoef(newy, c)
    correlation_xy = correlation_matrix[0, 1]
    r_squared = correlation_xy**2
    
    print(np.exp(popt), r_squared)
    return (newx, newy, newyerr, (np.exp(popt[0]), popt[1]), r_squared)

def getCoefCombined(x, y, level, yerr, sZPos = 98.92301, sYPos = 44.92301): 

    def func(x, a, b, c, d):
        # 44.92301 , 98.92301
        r = np.sqrt(sYPos ** 2 + (sZPos - x) ** 2)
        # return a * ((np.exp(b*r)) / (r**2) ) + c
        # return a * ((np.exp(b*r)) / (r**2) ) + c * np.exp(d*x) + e
        return a * x**3 + b * x**2 + c * x**1 + d

    # Ground
    if level == 1:
        newx = x[:14]
        newy = y[:14]
        newyerr = yerr[:14]
        
    # under ground
    elif level == 2:
        newx = x[9:]
        newy = y[9:]
        newyerr = yerr[9:]
        
    # total
    elif level == 3:
        newx = x
        newy = y
        newyerr = yerr
        
    # p0 = (100, 0.05, 0.008, 0.05)
    # p0 = (100, 0.001, 0.05)
    # p0 = (100, -1.0, 50, 0.1, 1)
    p0 = (100, -1.0, 50, 0.1)
    popt, pcov = curve_fit(func, newx, newy, p0=p0, method="lm", maxfev=100000)
    
    # get r squared
    c = func(newx, *popt)
    correlation_matrix = np.corrcoef(newy, c)
    correlation_xy = correlation_matrix[0, 1]
    r_squared = correlation_xy**2
    
    print(*popt, "\t R^2: ",r_squared)
    return (newx, newy, newyerr, popt, r_squared)
    


def getCoefCombinedPDAP(x, y, level, yerr, sZPos = 98.92301, sYPos = 44.92301):
    
    def func(x, a, b, c, d):
        # 44.92301 , 98.92301
        r = np.sqrt(sYPos ** 2 + (sZPos - x) ** 2)
        # return a * np.exp(b*r) + c
        # return a * (1/r**2) + b * np.exp(c * x) + d
        # return a * ((np.exp(b*r)) / (r**2) ) + c * np.exp(d*x) + e
        return a * x**3 + b * x**2 + c * x**1 + d

    if level == 1:
        newx = x[5:15]
        # print(newx)
        newy = y[5:15]
        newyerr = yerr[5:15]
        
    elif level == 2:
        newx = x[9:]
        newy = y[9:]
        newyerr = yerr[9:]
        
    elif level == 3:
        newx = x
        newy = y
        newyerr = yerr
        
    # p0 = (100, 0.05, 0.008, 0.05)
    # p0 = (100, -1.0, 50, 0.1, 1)
    p0 = (100, -1.0, 50, 0.1)
    popt, pcov = curve_fit(func, newx, newy, p0=p0, method="lm", maxfev=10000)
    
    # get r squared
    r = np.sqrt(sYPos ** 2 + (sZPos - newx) ** 2)
    c = func(newx, *popt)
    correlation_matrix = np.corrcoef(newy, c)
    correlation_xy = correlation_matrix[0, 1]
    r_squared = correlation_xy**2
    
    print(*popt, "\t R^2: ",r_squared)
    return (newx, newy, newyerr, popt, r_squared)
```
